```python
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
import pandas as pd
import random

from sklearn import preprocessing
```

```python
# load data set
dataset = pd.read_csv (r'D:\COURSES\Winter 2021 courses\Intelligent Analytics\HW2\Reactive-Ion-DSt.csv')
dataset.head()
```

```python
# Reading Feature inputs
X = dataset.iloc[:,1:7]
X.head()
```

```python
# Reading Labels outputs
Y = dataset.iloc[:,8:]
Y.head()
```

```python
# model building
model = Sequential()
model.add(Dense(16, input_dim=6, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(4, activation='linear'))
model.summary()
```

```python
# Compile model
model.compile(loss='mean_absolute_error', optimizer='adam', metrics=['mean_squared_error'])
```

```python
# Fit the model
model.fit(X, Y, epochs=150, batch_size=10,verbose = 1, validation_split = 0.2)
```
